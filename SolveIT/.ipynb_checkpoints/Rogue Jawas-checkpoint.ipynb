```python
incode = range(1,101)
outcode = []

for i in incode:
    temp = []
    if i % 5 == 0:
        temp.append('JAWA')
    if '5' in str(i):
        temp.append('JAWA')
    if i % 7 == 0:
        temp.append('BEANS')
    if '7' in str(i):
        temp.append('BEAN5')
    if not temp:
        temp.append(str(i))
    tempstr = " ".join(temp)
    outcode.append(tempstr)

for line in outcode:
    print line
```
